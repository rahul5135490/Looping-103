```python
print("i,j,k,l,m")
for i in range(5):
    for j in range(5):
        for k in range(5):
            for l in range(5):
                for m in range(5):
                    print(i,j,k,l,m)
```

```python
for i in range(5):
    for j in range(5):
        if(j<=i):
            print("*",end="")
        else:
            print("+",end="")
    print(" ")
```
